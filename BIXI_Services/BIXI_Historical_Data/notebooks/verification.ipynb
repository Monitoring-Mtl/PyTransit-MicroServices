```python
import pandas as pd
import time
import requests
```

```python
raw_data = pd.read_csv("2024/DonneesOuvertes2024_0102.csv", sep=',')
df = pd.DataFrame(raw_data)
df = df.dropna()
df = df[['STARTSTATIONNAME','ENDSTATIONNAME','STARTTIMEMS','ENDTIMEMS']]
df['STARTTIMEMS']=pd.to_datetime(df['STARTTIMEMS'], unit='ms')
df['ENDTIMEMS']=pd.to_datetime(df['ENDTIMEMS'], unit='ms')
df['YEAR_E'] = df['ENDTIMEMS'].dt.year
df['MONTH_E'] = df['ENDTIMEMS'].dt.month
df['DAY_E'] = df['ENDTIMEMS'].dt.day
df['HOUR_E'] = df['ENDTIMEMS'].dt.hour
df['MIN_E'] = df['ENDTIMEMS'].dt.minute
df['YEAR_s'] = df['STARTTIMEMS'].dt.year
df['MONTH_S'] = df['STARTTIMEMS'].dt.month
df['DAY_S'] = df['STARTTIMEMS'].dt.day
df['HOUR_s'] = df['STARTTIMEMS'].dt.hour
df['MIN_S'] = df['STARTTIMEMS'].dt.minute
df
```

```python
df.to_csv('Clean-DonneesOuvertes2023.csv', index=False)
```

```python
#count number of trips between two stations and add value as new column 
nb_trips = pd.DataFrame(df.value_counts(['STARTSTATIONNAME','ENDSTATIONNAME'])).reset_index()
df_rank = df.merge(nb_trips, left_on = ['STARTSTATIONNAME','ENDSTATIONNAME'], right_on= ['STARTSTATIONNAME','ENDSTATIONNAME'])     
df_rank.sort_values(by='STARTSTATIONNAME',inplace=True)
```

```python
df_rank[(df_rank['STARTSTATIONNAME']=='Peel / Notre-Dame')&(df_rank['ENDSTATIONNAME']=='Place du Canada (Peel / des Canadiens-de-Montréal)')]
```

```python
#calculate the time each trip took
df_rank["DELTATIME"] = df_rank["ENDTIMEMS"]-df_rank["STARTTIMEMS"] 
df_rank['DELTATIME'] = pd.to_timedelta(df_rank['DELTATIME'], unit='ms')
df_rank.sort_values('count',ascending=False, inplace=True)
mean_df= df_rank.groupby(['STARTSTATIONNAME','ENDSTATIONNAME','count'],as_index=False)['DELTATIME'].mean()
mean_df.rename(columns={'count':'NUMBEROFTRIPS'}, inplace=True)
mean_df.sort_values('DELTATIME',ascending=False, inplace=True)
mean_df
```

```python
notre_dame=mean_df[(mean_df['STARTSTATIONNAME']=='Peel / Notre-Dame')&(mean_df['ENDSTATIONNAME']=='Place du Canada (Peel / des Canadiens-de-Montréal)')]
notre_dame.sort_values(by="NUMBEROFTRIPS", ascending= False)
```
